DEVELOP A SAMPLING TECHNIQUE

```r
skewness(bh)
```

SAMPLING THE DATASET

```r
set.seed(120)
```

```r
stats = data.frame(feature=colnames(bh))
stats['mean_'] = sapply(bh, mean)
stats['sd_'] = sapply(bh, sd)
stats['skewness_'] = sapply(bh, skewness)
stats
```

```r
bh_features = Filter(is.numeric, bh)
```

```r
set.seed(102)
```

```r
bh05 <- bh[sample(1:nrow(bh), 05, replace = FALSE),]
bh20 <- bh[sample(1:nrow(bh), 20, replace = FALSE),]
bh50 <- bh[sample(1:nrow(bh), 50, replace = FALSE),]
bh100 <- bh[sample(1:nrow(bh), 100, replace = FALSE),]
```

```r
bh05
bh20
bh50
bh100
```

lets evaluate for n = 5, 

```r
stats05 = data.frame(feature=colnames(bh05))
stats05['mean_'] = sapply(bh05, mean)
stats05['sd_'] = sapply(bh05, sd)
stats05['skewness_'] = sapply(bh05, skewness)
stats05
```

COMPARE TO THE ACTUAL MEAN

```r
colMeans(bh05) - stats$mean_
```

AVERAGE DISTANCE TO THE MEAN,

```r
((colMeans(bh05) - stats$mean)/stats$sd) 
```

AS WE CAN SEE FOR N = 5, WE ARE FAR FROM THE MEAN.
LETS TRY IT FOR N = 20

Type Markdown and LaTeX:  $α2$ 


Repeatedly Sample

```r
stats20 = data.frame(feature=colnames(bh20))
stats20['mean'] = sapply(bh20, mean)
stats20['sd'] = sapply(bh20, sd)
stats20['skewness'] = sapply(bh20, skewness)
stats20
```

COMPARE TO THE MEAN

```r
colMeans(bh20) - stats$mean
```

```r
AVERAGE DISTANCE TO THE MEAN,
```

```r
((colMeans(bh20) - stats$mean)/stats$sd) 
```

we are still far from the Mean
lets try N = 50

```r
stats50 = data.frame(feature=colnames(bh50))
stats50['mean'] = sapply(bh50, mean)
stats50['sd'] = sapply(bh50, sd)
stats50['skewness'] = sapply(bh50, skewness)
stats50
```

```r
colMeans(bh50) - stats$mean
```

```r
((colMeans(bh50) - stats$mean)/stats$sd) 
```

AS WE SAMPLE MORE THE N = 50, THE DISTANCE IS GETTING SMALL
NOW LETS TRY FOR N = 100

```r
stats100 = data.frame(feature=colnames(bh100))
stats100['mean'] = sapply(bh100, mean)
stats100['sd'] = sapply(bh100, sd)
stats100['skewness'] = sapply(bh100, skewness)
stats100
```

```r
colMeans(bh100) - stats$mean
```

THE DISTANCE TO THE MEAN

```r
((colMeans(bh100) - stats$mean)/stats$sd) 
```

the distance is very small, near 5%, as the sample is big, the mean for N = 100 is close to the actual mean

As we can see, as we perform more times, the distance between the mean and the standart deviation is getting small. the mean for 100 times is very close to the mean of the data set. n= 100 is a good sample size to work with.
